```python
from lazypredict.Supervised import LazyRegressor, LazyClassifier
```

```python
def transform(value):
  categories = [0,0.5,1,2,3]
  differences = [abs(x-value) for x in categories]
  min_index = differences.index(min(differences))
  return str(categories[min_index])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# read the data
data = pd.read_csv("final_data.csv")

# Creating train test split
y = data['Target']
print(type(y))
X = data.drop(columns=['Target'], axis=1)
y = y.apply(transform)

#Call train_test_split on the data and capture the results
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
classi=LazyClassifier(verbose=0,predictions=True)
models_c,predictions_c=classi.fit(X_train, X_test, y_train, y_test)
```

```python
models_c.drop(['ROC AUC','Balanced Accuracy'],axis=1)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
# read the data
data = pd.read_csv("final_data.csv")

# Creating train test split
y = data['Target']
#print(type(y))
X = data.drop(columns=['Target'], axis=1)
y = y.apply(transform)

smote = SMOTE(k_neighbors=3)

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(X, y)

print('Original dataset shape', Counter(y))
print('Resample dataset shape', Counter(y_smote))
```

```python
up_sampled_data = x_smote.copy()
up_sampled_data['Target'] = y_smote 
up_sampled_data.head()
```
